**Step 1: Importing the libraries**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler,StandardScaler,OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
```

**Step 2: Importing dataset**

```python
df=pd.read_csv('/content/drive/MyDrive/Task7/Assignment11_to_16/Data.csv')
df
```

```python
df.info()
```

**Step 3: Handling the missing data**

```python
df.isnull().sum()
```

```python
Imp=SimpleImputer(missing_values=np.nan,strategy='mean')
df.Age=Imp.fit_transform(df['Age'].values.reshape(-1,1))[:,0]
df.Salary=Imp.fit_transform(df['Salary'].values.reshape(-1,1))[:,0]
print(df)
```

**Step 4: Encoding categorical data**

```python
oe=OrdinalEncoder()
df1=pd.DataFrame(oe.fit_transform(df[['Country','Purchased']])).astype('int')
df1.columns=['Country','Purchased']
df1
```

```python
df=df.drop(['Country','Purchased'],axis=1)
```

```python
df=df.astype('int')
df=df.join(df1)
df
```

**Step 5: Creating a dummy variable**

```python
new_df=pd.get_dummies(data=df,columns=['Country','Purchased'])
new_df.columns=['Age', 'Salary', 'Country_France', 'Country_Germany', 'Country_Spain', 'Purchased_No','Purchased_Yes']
new_df
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
x=df[['Age','Salary']]
y=df['Purchased']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

**Step 7: Feature Scaling**

```python
mscaler=MinMaxScaler()
x_train=mscaler.fit_transform(x_train)
x_train
```
